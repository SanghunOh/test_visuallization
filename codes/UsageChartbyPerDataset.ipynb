* 다변수 (범주/연속) : 가장 빈번한 방식

```python
sns.barplot(data=df1, x='channel',y='amount') # 평균 출력
```

```python
df1['channel'].value_counts()
```

```python
# 표시 많다고 amount 순서로 줄이면 않됨.
p1 = pd.pivot_table(data=df1, index='channel', values='amount', aggfunc='sum')
p1.sort_values(by='amount', ascending=False).head(10)
```

```python
top_list = p1.sort_values(by='amount', ascending=False).head(5).index.tolist()
```

```python
sns.barplot(data=df1, x='channel',y='amount', order=top_list) # 순서 출력
```

```python
sns.barplot(data=df1, x='channel',y='amount', order=top_list, hue='type_of_contract') # 순서 출력
```

```python
sns.barplot(data=df1, x='product',y='age')
```

```python
sns.barplot(data=df1, x='product',y='age',ci=None) # 신뢰구간(confidence interval) None
```

```python
sns.barplot(data=df1, x='product',y='amount',ci=None,estimator=sum) # 합계
```

```python
sns.boxplot(data=df1, x='channel',y='amount', order=top_list) # 순서 출력
```

* 다변수 (시간/연속)
- 시간 다루기 : 문자 / 숫자로 된 데이터 다루기

```python
df1['datetime(timestmp)'] = (pd.to_datetime(df1['datetime']))
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount')
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount',ci=None)
```

```python
sns.lineplot(data=df1,x='datetime(timestmp)',y='amount',ci=None,estimator=sum)
```

```python
df1['계약요일'] = df1['datetime(timestmp)'].dt.day_name()
```

```python
sns.barplot(data=df1,x='계약요일',y='amount',ci=None,estimator=sum)
```

```python
df1['계약주차'] = df1['datetime(timestmp)'].dt.weekday
```

```python
sns.lineplot(data=df1,x='계약주차',y='amount',ci=None)
```

```python
df1['datetime(timestmp)'].describe()
```

#### 해결 방안 : 년도 주차 세분화 or 머지화

```python
df1['계약년도'] = df1['datetime(timestmp)'].dt.year
```

```python
sns.pointplot(data=df1,x='계약주차',y='amount',ci=None,estimator=sum, hue='계약년도')
```

```python
cond1 = (df1['계약년도'] != 2019)
df2 = df1.loc[cond1]
sns.pointplot(data=df2,x='계약주차',y='amount',ci=None,estimator=sum)
```
