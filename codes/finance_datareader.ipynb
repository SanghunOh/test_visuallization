```python
import FinanceDataReader as fdr

# Get the daily price data for Apple (AAPL) between 2020-01-01 and 2020-12-31
aapl_df = fdr.DataReader('AAPL', '2020-01-01', '2020-12-31')

# Print the first 5 rows of the data frame
print(aapl_df.head())

# Get the monthly price data for Tesla (TSLA) between 2010-01-01 and 2021-01-31
tsla_df = fdr.DataReader('TSLA', '2020-01-01', '2020-12-31')

# Print the first 5 rows of the data frame
print(tsla_df.head())

# Get the ticker list for stocks from KOSPI market in Korea
kospi_tickers = fdr.StockListing('KOSPI')

# Print the first 5 tickers
print(kospi_tickers.head())

# Get the ticker list for ETFs from Korea Exchange
etf_tickers = fdr.StockListing("ETF/KR")

# Print the first 5 tickers
print(etf_tickers.head())
```

```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

# Get the daily price data for Samsung Electronics (005930) between 2020-01-01 and 2020-12-31
samsung_df = fdr.DataReader('005930', '2020-01-01', '2020-12-31')
lgenergy_df = fdr.DataReader('066570', '2020-01-01', '2020-12-31')

# Plot the closing prices as a line chart
plt.plot(samsung_df.index, samsung_df['Close'], color='red')
plt.plot(lgenergy_df.index, lgenergy_df['Close'], color='green')
plt.title('Daily Closing Prices in 2020')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.show()
```
