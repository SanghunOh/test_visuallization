```
import folium
```

```
m = folium.Map(
    location=[45.5236, -122.6750],
    # tiles='Stamen Toner',
    # tiles='Stamen Terrain'
    zoom_start=15
)
m
```

```
tooltip = 'Click me!'

folium.Marker([45.5236, -122.6750], popup='<i>Mt. Hood Meadows</i>', tooltip=tooltip).add_to(m)
folium.Marker([45.5236, -122.6790], popup='<i>Move Left</i>', tooltip=tooltip).add_to(m)
folium.Marker([45.5236, -122.6710], popup='<b>Move Right</b>', tooltip=tooltip).add_to(m)
folium.Marker([45.5276, -122.6750], popup='<i>Move Top</i>', tooltip=tooltip).add_to(m)
folium.Marker([45.5136, -122.6750], popup='<i>Move Bottom</i>', tooltip=tooltip).add_to(m)
m
```

```
folium.Marker(
    location=[45.5336, -122.6850],
    popup='Timberline Lodge',
    icon=folium.Icon(color='green')
).add_to(m)
m
```

```
folium.CircleMarker(
    location=[45.5276, -122.6600],
    radius=20,
    popup='Laurelhurst Park',
    color='red',       # color='#3186cc',
    fill=True,
    fill_color='green'
).add_to(m)
m
```
