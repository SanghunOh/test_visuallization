##### Line Chart

```python
# plt.figure(figsize=(3,3))
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x, y)
```

```python
plt.figure(figsize=(4,3))

x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x1,y1)
plt.plot(x,y)

plt.savefig('test_img.jpg')
```

```python
plt.figure(figsize=(4,3))
x1 = range(20)
y1 = [np.cos(value) for value in x1]

x2 = range(20)
y2 = [np.sin(value) for value in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize=(4,3))

plt.subplot(1,2,1)
plt.plot(x1,y1, color='red', linestyle='dotted')

plt.subplot(1,2,2)
plt.plot(x2,y2, color='blue', linestyle='dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1, y1, c='#a52a2a', ls='-.', label='Line 1')
plt.plot(x2, y2, c='#0000ff', ls=':', label='Line 2')

plt.text(50,60, 'Line 1')
plt.annotate('Line 2', xytext=(30,110), xy=(50,100),
            arrowprops = dict(facecolor='yellow', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc='upper right')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line 2')
```

#### Scatter Plot

```python
plt.figure(figsize=(4,3))
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.scatter(x,y)
```

```python
x = np.random.rand(100)
y = x + np.random.randn(100)/10

plt.scatter(x,y)
```

```python
data1 = np.random.rand(512, 2)
data2 = np.random.rand(512, 2)

plt.scatter(data1[:,0], data1[:, 1], c='b', marker='x')
plt.scatter(data2[:,0], data1[:, 1], c='r', marker='o')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)
area = np.pi*(30*np.random.rand(50))**2

plt.scatter(x, y, c=colors, s = area, marker='o', alpha=0.3)
```
